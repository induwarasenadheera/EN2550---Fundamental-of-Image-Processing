```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im=cv.imread(r'./Images/spider.png',cv.IMREAD_GRAYSCALE)
assert cv is not None

gamma=[0.2,0.8,1.2,2]

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',im)
cv.waitKey(0)

fig,ax=plt.subplots()
plt.imshow(im,'gray')
plt.title('Origial Grayscale Image')

fig,ax=plt.subplots(2,2,figsize=(12,10))

for _ in range(4):
    t=np.array([(p/255)**gamma[_]*255 for p in range(0,256)]).astype(np.uint8)
    g=cv.LUT(im,t)#look up table

    plt.subplot(2,2,_+1)
    plt.imshow(g,'gray')
    stri='gamma '+str(gamma[_])
    plt.title(stri)

    cv.waitKey(0)
    cv.imshow('Image',g)
    cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f=cv.imread(r'./Images/shells.tif',cv.IMREAD_GRAYSCALE)
assert f is not None

hist_f=cv.calcHist([f],[0],None,[256],[0,256])
g=cv.equalizeHist(f)
hist_g=cv.calcHist([g],[0],None,[256],[0,256])

fig,ax=plt.subplots(2,1,figsize=(8,8))
ax[0].plot(hist_f)
ax[1].plot(hist_g)

fig,ax=plt.subplots(2,1,figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(f,"gray")
plt.title('Original image')
plt.subplot(1,2,2)
plt.title('Histogram equalize image')
plt.imshow(g,"gray")

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f=cv.imread(r'./Images/zion_pass.jpg')
assert f is not None
f_=cv.cvtColor(f,cv.COLOR_BGR2HSV)
h, s, v = cv.split(f_)

t1=np.linspace(0,100,50)
t2=np.linspace(100,255,150)
t3=np.linspace(255,255,56   )

t=np.concatenate((t1,t2,t3),axis=0).astype(np.uint8)

fig,ax=plt.subplots()
assert len(t) ==256
plt.plot(t)
plt.title("Saturation Transform")
ss=cv.LUT(s,t)#look up table

final_hsv = cv.merge((h, ss, v))
img2 = cv.cvtColor(final_hsv, cv.COLOR_HSV2BGR)

fig,ax=plt.subplots(2,1,figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
plt.title('Original image')
plt.subplot(1,2,2)
plt.title('Saturation Changed image')
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))

cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
cv.imshow('Image', f)
cv.waitKey(0)
cv.imshow('Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f=cv.imread(r'./Images/zion_pass.jpg')
assert f is not None
f_=cv.cvtColor(f,cv.COLOR_BGR2HSV)
h, s, v = cv.split(f_)

t1=np.linspace(0,100,50)
t2=np.linspace(100,255,150)
t3=np.linspace(255,255,56)

t=np.concatenate((t1,t2,t3),axis=0).astype(np.uint8)

fig,ax=plt.subplots()
assert len(t) ==256
plt.plot(t)
plt.title("Saturation Transform")
ss=cv.LUT(s,t)#look up table

n1=np.linspace(0,110,90)
n2=np.linspace(110,179,90)
n3=np.linspace(255,255,76)

n=np.concatenate((n1,n2,n3),axis=0).astype(np.uint8)

fig,ax=plt.subplots()
assert len(n) ==256
plt.plot(n)
plt.title("Hue Transform")
hh=cv.LUT(h,n)#look up table

final_hsv = cv.merge((hh, ss, v))
img2 = cv.cvtColor(final_hsv, cv.COLOR_HSV2BGR)

fig,ax=plt.subplots(2,1,figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
plt.title('Original image')
plt.subplot(1,2,2)
plt.title('Saturation Changed image')
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))

cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
cv.imshow('Image', f)
cv.waitKey(0)
cv.imshow('Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()



```
