```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im=cv.imread(r'./Images/Lenna.png',cv.IMREAD_GRAYSCALE)
assert cv is not None

gamma=0.5
t=np.array([(p/255)**gamma*255 for p in range(0,256)]).astype(np.uint8)
g=cv.LUT(im,t)#look up table
#g=t[cv]

fig,ax=plt.subplots()
ax.plot(t)


cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',im)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im=cv.imread(r'./Images/spider.png',cv.IMREAD_GRAYSCALE)
assert cv is not None

gamma=[0.2,0.8,1.2,2]
cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',im)
fig,ax=plt.subplots()
plt.imshow(im,'gray')
fig,ax=plt.subplots(2,2,figsize=(12,10))
for _ in range(4):
    t=np.array([(p/255)**gamma[_]*255 for p in range(0,256)]).astype(np.uint8)
    g=cv.LUT(im,t)#look up table
    #g=t[cv]
    plt.subplot(2,2,_+1)
    plt.imshow(g,'gray')
    stri='gamma '+str(gamma[_])
    plt.title(stri)

    cv.waitKey(0)
    cv.imshow('Image',g)
    cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f=cv.imread(r'./Images/shells.tif',cv.IMREAD_GRAYSCALE)
assert f is not None

hist_f=cv.calcHist([f],[0],None,[256],[0,256])
g=cv.equalizeHist(f)
hist_g=cv.calcHist([g],[0],None,[256],[0,256])

fig,ax=plt.subplots(2,1,figsize=(8,8))
ax[0].plot(hist_f)
ax[1].plot(hist_g)

fig,ax=plt.subplots(2,1,figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(f,"gray")
plt.title('Original image')
plt.subplot(1,2,2)
plt.title('Histogram equalize image')
plt.imshow(g,"gray")

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()
```
